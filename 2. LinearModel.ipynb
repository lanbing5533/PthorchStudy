```python
# Linear model by python3
import numpy as np
import matplotlib.pyplot as plt 

# training set
x = [1.0, 2.0, 3.0]
y = [4.0, 5.0, 6.0]

# linear model: y = w*x
def forward(x):
    return x * w

# loss = (y^ - y)**2
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

w_list = [] # save the w
mse_list = [] # cost value of each w

# cost value at [0.0, 0.1, 0.2 ,..., 4.0]
for w in np.arange(0.0, 4.1, 0.1):
    print('w=', w)
    l_sum = 0
    # For each sample in train set, the loss function values were computed.
    for x_val, y_val in zip(x, y):
        y_pred_val = forward(x_val)
        loss_val = loss(x_val, y_val)
        # ATTENTION: Value of cost function is the sum of loss function.
        l_sum += loss_val
        print('\t:', x_val, y_val, y_pred_val, loss_val)
    print('MSB=', l_sum / 3)
    w_list.append(w)
    mse_list.append(l_sum / 3)

# draw the graph
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
